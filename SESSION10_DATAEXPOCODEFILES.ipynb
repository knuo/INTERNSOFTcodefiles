#Plotting in pandas

```python
norm = pd.Series(np.random.normal(size=50))
norm.plot()
print(norm)
```

```python
norm.cumsum().plot(grid=True)
```

```python
v = pd.DataFrame({'normal':np.random.normal(size=15),'gamma':np.random.gamma(1,size=15),'poisson':np.random.poisson(size=15)})
print(v)
v.cumsum(0).plot()
```

```python
v.cumsum(0).plot(subplots=True)
```

```python
v.cumsum(0).plot(secondary_y='poisson')                                                                                                                                                             
```

```python
fig, axes = plt.subplots(nrows=3,ncols=1,figsize=(4,12))
for i,var in enumerate(['normal','gamma','poisson']):
    v[var].cumsum(0).plot(ax=axes[i],title=var)
axes[0].set_ylabel('cumulative sum')
```

#simple line plots

```python
fig = plt.figure()
ax = plt.axes()
s= np.linspace(0,30,60, endpoint=True)
print(s)
ax.plot(s,np.sin(s))
```

```python
plt.plot(s,np.sin(s))
```

#line colors and styles

```python
plt.plot(s, np.sin(s-0),color='red')
plt.plot(s, np.sin(s-1),color='y')
plt.plot(s, np.sin(s-2),color='0.33')
```

```python
plt.plot(s, np.sin(s-0),linestyle='-')
plt.plot(s, np.sin(s-1),linestyle=':')
plt.plot(s, np.sin(s-2),linestyle='-.')
```

```python
plt.plot(s, np.sin(s-0))
plt.xlim(-1,15)
plt.ylim(-2,2)
```

#labeling plots

```python
plt.plot(s,np.cos(s))
plt.title("cosine curve")
plt.xlabel('x')
plt.ylabel('cos(x))')
```

```python
abc = np.random.RandomState(7)
x = abc.randn(100)
y = abc.randn(100)
colors = abc.rand(100)
sizes= 1000*abc.rand(100)
 
plt.scatter(x,y, c= colors,s=sizes,alpha=0.5,cmap='Reds')
plt.colorbar()
```

#Errorbars

```python
x = np.linspace(0,10,50)
dy = 0.1
dx = 0.2
y = np.sin(x)+ dy *np.random.randn(50)
plt.errorbar(x,y,xerr=dx,yerr=dy,fmt='.r')
```

#Histograms

```python
plt.style.use('seaborn-white')
data = np.random.randn(10000)
print(data)
```

```python
plt.hist(data)
```

```python
#normed kwargs was deprecated, using density instead
plt.hist(data,bins=30, density = True, alpha=0.6,histtype='stepfilled')
```

```python
x = np.random.normal(0,0.8,100)
y = np.random.normal(-2,1,100)
 
kwargs = dict(histtype='stepfilled',alpha=0.5,density = True,bins=20)
plt.hist(x,**kwargs)
plt.hist(y,**kwargs)
```

```python
counts,bins_edges = np.histogram(data,bins=5)
counts
```

#2D histograms

```python
mean = [5,2]
cov= [[3,1],[2,0.5]]
x,y = np.random.multivariate_normal(mean,cov,2)
```

```python
plt.hist2d(x,y, bins=20,cmap='Reds')
cb= plt.colorbar()
```

```python
plt.hexbin(x,y,gridsize=20,cmap='Reds')
cb = plt.colorbar()
```
